```python
import numpy as np
import matplotlib.pyplot as plt

def cosine_wave(x):
    return np.cos(x)

x_values = np.linspace(0, 2*np.pi,420)
y_values = cosine_wave(x_values)

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, color='cyan', label='cos(x)', linewidth=2)

plt.title('Cos Wave')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(color='black', linestyle='-', linewidth=0.5)
plt.gca().set_facecolor('black')
plt.show()
     
```
