```python
import numpy as np
import matplotlib.pyplot as plt

def sine_wave(x):
    return np.sin(x)

x_values = np.linspace(0, 2*np.pi,360)
y_values = sine_wave(x_values)

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, color='cyan', label='sin(x)', linewidth=2)

plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(color='black', linestyle='-', linewidth=0.5)
plt.gca().set_facecolor('black')
plt.show()
```
